```python
from collections import defaultdict
import json

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
teams = [
    'PHI', 'ATL', 'BUF', 'BAL', 'CLE', 'PIT', 'IND', 'CIN', 'MIA',
    'TEN', 'SF', 'MIN', 'HOU', 'NE', 'TB', 'NO', 'NYG', 'JAX', 'KC',
    'LAC', 'ARI', 'WAS', 'CAR', 'DAL', 'SEA', 'DEN', 'CHI', 'GB',
    'DET', 'NYJ', 'LA', 'OAK', 'JAC', 'SD', 'STL'
]
```

```python
def preprocess_drives(drives):
    """Preprocess drives for analysis."""
    df = pd.DataFrame(drives)
    df['drive_id'] = df.index
    df['points'] = 0
    df.loc[df['result'] == 'Touchdown', 'points'] = 7
    df.loc[df['result'] == 'Field Goal', 'points'] = 3
    df['total_yards'] = df['penalty_yards'] + df['yards_gained']
    lower = np.arange(0, 100, 5)
    upper = np.arange(5, 105, 5)
    bins_list = list(zip(lower, upper))
    bins = pd.IntervalIndex.from_tuples(bins_list)
    df['yard_line_bin'] = pd.cut(df['start_yard_line'], bins)
    df['yard_line_bin'] = df['yard_line_bin'].map(lambda x: '%s-%s' % (x.left, x.right))
    return df
```

```python
df = pd.DataFrame()

for season in range(2009, 2019):
    drives = json.load(open('./data/%i_drives.json' % season, 'r'))
    sdf = preprocess_drives(drives)
    sdf['season'] = season
    df = pd.concat((df, sdf))
```

```python
all_agg = df.groupby('yard_line_bin').agg({'drive_id': 'count', 'points': 'mean'})
```

```python
kc18 = df.loc[(df['offensive_team'] == 'KC') & (df['season'] == 2018)].copy()
kc18.head()
```

```python
kc_agg = kc18.groupby('yard_line_bin').agg({'drive_id': 'count', 'points': 'mean'})
```

```python
comp = all_agg.join(kc_agg, lsuffix='_nfl', rsuffix='_kc')
```

```python
comp['weight'] = comp['drive_id_kc'] / comp['drive_id_kc'].sum()
```

```python
comp['owp'] = (comp['points_kc'] - comp['points_nfl']) * comp['weight']
```

```python
import numpy as np
import pandas as pd

from score_teams import score_drives
from score_teams import aggregate_game_drives, aggregate_season_drives
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gdf = aggregate_game_drives(2009, 2018, exclude_playoffs=True, exclude_blowouts=14)

gdf.head()
```

```python
gdf['adj_offensive_score_copy'] = gdf['adj_offensive_score'].copy()

sdf = gdf.groupby(['season', 'offensive_team'], as_index=False).agg(
    {'adj_offensive_score': 'mean', 'adj_offensive_score_copy': 'std'}
)

sdf = sdf.rename({'adj_offensive_score_copy': 'adj_offensive_score_std'}, axis=1)
sdf = sdf.sort_values('adj_offensive_score', ascending=False)

sdf.head(20)
```

```python
columns_a = [col for col in sdf.columns if col != 'season_lag']

sdf['season_lag'] = sdf['season'] - 1 

columns_b = ['season_lag'] + [col for col in sdf.columns if 'season' not in col]

mdf = sdf[columns_a].merge(
        sdf[columns_b], left_on=['season', 'offensive_team'], 
        right_on=['season_lag', 'offensive_team'], suffixes=('', '_next')
)

mdf = mdf.sort_values('adj_offensive_score', ascending=False)

mdf['performance_delta'] = mdf['adj_offensive_score_next'] - mdf['adj_offensive_score']
```
